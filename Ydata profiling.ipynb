```python
import pandas as pd
from ydata_profiling import ProfileReport
```

# Load the data

```python
df = pd.read_csv("sample_dataset.csv")
```

# Produce the full profiling report

```python
Full_profile = ProfileReport(df,title="Full Sample_dataset Report")
```

# Saving the full profiling report

```python
Full_profile.to_file("FullFull Sample_dataset Report.html")
```

# Producing and saving first five features report

```python
First_Five_Features_report = ProfileReport(df.iloc[:,0:5])
```

```python
First_Five_Features_report.to_notebook_iframe()
```

```python
First_Five_Features_report.to_file("First_Five_Features_report.html",silent=False)
```

# Avoid the calculation of multivariate statistics

```python
First_Five_Features_report = ProfileReport(df.iloc[:,0:5],minimal=True)
First_Five_Features_report.to_notebook_iframe()
```
